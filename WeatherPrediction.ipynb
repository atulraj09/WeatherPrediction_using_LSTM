```python
temp = data['Temp_C']
temp.plot()
```

```python
def data_to_X_y(data,window_size=5):
  data_as_np = data.to_numpy()
  X = []
  y = []
  for i in range(len(data_as_np)-window_size):
    row = [[a] for a in data_as_np[i:i+5]]
    X.append(row)
    label = data_as_np[i+5]
    y.append(label)
  return np.array(X),np.array(y)
```

```python
WINDOW_SIZE = 5
X, y = data_to_X_y(temp,WINDOW_SIZE)
X.shape,y.shape
```

```python
X_train,y_train = X[:7000],y[:7000]
X_val,y_val = X[7000:7500],y[7000:7500]
X_test,y_test = X[7500:],y[7500:]
X_train.shape,y_train.shape,X_val.shape,y_val.shape,X_test.shape,y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

model1 = Sequential()
model1.add(InputLayer((5,1)))
model1.add(LSTM(64))
model1.add(Dense(8,'relu'))
model1.add(Dense(1,'linear'))

model1.summary()
```

```python
cp = ModelCheckpoint('model1/.keras',save_best_only=True)
model1.compile(loss=MeanSquaredError(),optimizer = Adam(learning_rate=0.0001),metrics = [RootMeanSquaredError()])
```

```python
model1.fit(X_train,y_train,validation_data = (X_val,y_val),epochs = 10, callbacks = [cp])
```

```python
from tensorflow.keras.models import load_model
model1 = load_model('model1/.keras')
```

### **Model Optimization using Approximate Bayesian Computation**


```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, InputLayer
from sklearn.metrics import mean_squared_error
import random

# Define the LSTM model architecture
def build_lstm(hidden_units):
    model = Sequential()
    model.add(InputLayer((5, 1)))  # Input shape: (time_steps, features)
    model.add(LSTM(hidden_units))
    model.add(Dense(8, activation='relu'))
    model.add(Dense(1, activation='linear'))
    return model
```

```python
# Define the prior distribution
def sample_prior():
    return {
        "hidden_units": random.randint(32, 128),  # Number of LSTM units
        "learning_rate": np.random.uniform(0.0001, 0.01)  # Learning rate
    }
```

```python
# Simulate the model and calculate a distance metric
def simulate_and_score(params, X_train, y_train, X_val, y_val):
    # Build model with sampled parameters
    model = build_lstm(params["hidden_units"])
    model.compile(optimizer="adam", loss="mse", metrics=["mse"])

    # Use only a few epochs for efficiency in ABC
    model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=5, verbose=0)

    # Predict on validation set
    val_predictions = model.predict(X_val).flatten()

    # Calculate RMSE as the distance metric
    mse = mean_squared_error(y_val, val_predictions)
    rmse = np.sqrt(mse)
    return rmse, model
```

```python
def abc_optimization(X_train, y_train, X_val, y_val, num_samples, epsilon):
    accepted_params = []
    accepted_models = []
    distances = []

    for i in range(num_samples):
        # Sample parameters from the prior
        params = sample_prior()

        # Simulate and calculate distance
        distance, model = simulate_and_score(params, X_train, y_train, X_val, y_val)

        # Accept parameters if distance is below the threshold epsilon
        if distance < epsilon:
            accepted_params.append(params)
            accepted_models.append(model)
            distances.append(distance)

        print(f"Iteration {i+1}/{num_samples}, Distance: {distance}, Accepted: {len(accepted_params)}")

    return accepted_params, accepted_models, distances
```

```python
# Run the ABC optimization
num_samples = 100  # Number of parameter samples
epsilon = 0.76  # Acceptance threshold for RMSE

accepted_params, accepted_models, distances = abc_optimization(
    X_train, y_train, X_val, y_val, num_samples, epsilon
)
```

```python
# Analyze the results
print("\nAccepted Parameters:")
for i, params in enumerate(accepted_params):
    print(f"Model {i+1}: {params}, Distance: {distances[i]}")
```

```python
import matplotlib.pyplot as plt

# Plot RMSE Distribution
plt.figure(figsize=(6, 4))
plt.hist(distances, bins=20, color='skyblue', edgecolor='black')
plt.title('RMSE Distribution', fontsize=14)
plt.xlabel('RMSE', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(alpha=0.4)
plt.show()
```

```python
# Extract Learning Rate and RMSE
learning_rates = [params['learning_rate'] for params in accepted_params]

# Scatter plot: Learning Rate vs RMSE
plt.figure(figsize=(6, 4))
plt.scatter(learning_rates, distances, color='orange', alpha=0.7)
plt.title('Learning Rate vs RMSE', fontsize=14)
plt.xlabel('Learning Rate', fontsize=12)
plt.ylabel('RMSE', fontsize=12)
plt.grid(alpha=0.4)
plt.show()
```

```python
# Extract Hidden Units and RMSE
hidden_units = [params['hidden_units'] for params in accepted_params]

# Scatter plot: Hidden Units vs RMSE
plt.figure(figsize=(6, 4))
plt.scatter(hidden_units, distances, color='green', alpha=0.7)
plt.title('Hidden Units vs RMSE', fontsize=14)
plt.xlabel('Hidden Units', fontsize=12)
plt.ylabel('RMSE', fontsize=12)
plt.grid(alpha=0.4)
plt.show()
```
